# Task 1: Write a dynamic function to find the derivative of any function f(x)

```python
import sympy as sp

def dy_dx(func, var):
  # Define the variable var as a symbol
  var = sp.Symbol(var)
  # Compute the derivative of the function func with respect to var
  derivative = sp.diff(func, var)
  return derivative
```

# Task 2: Test the derivative function written with an equation of your choice
*NB: Must have atleast 2 minimas and atleast 2 maximas*

```python
# Define the variable
x = sp.symbols('x')

# Equation : f(x) = x^5 - 10x^3 + 15x
f = x**5 - 10*x**3 + 15*x

# Compute the derivative of the function
derivative_f = dy_dx(f, 'x')
print(f"Derivative of f(x) = x^5 - 10x^3 + 15x=> {derivative_f}")
```

#Task 3: Plot a graph of the Equation must have 4 points (2 minimas and 2 maximas)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the variable and function
x = sp.symbols('x')
f = x**5 - 10*x**3 + 15*x

# Compute the derivative
derivative_f = sp.diff(f, x)

# Finding critical points
critical_points_f = sp.solve(derivative_f, x)

# Create an array of x values for plotting
x_vals = np.linspace(-3, 3, 400)
# Evaluate the function for these x values
f_vals = [f.subs(x, val) for val in x_vals]

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label=r'$f(x) = x^5 - 10x^3 + 15x$', color='blue')

# Mark critical points on the plot
for point in critical_points_f:
    y_val = f.subs(x, point)
    plt.scatter(float(point), float(y_val), color='red', zorder=5)
    plt.text(float(point), float(y_val), f'({float(point):.2f}, {float(y_val):.2f})', fontsize=9, ha='right')

# Add grid, labels, title, and legend
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.title('Graph of the Function with Critical Points')
plt.xlabel('x')
plt.ylabel('y or f(x)')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

# Task 4: Create separate array of Minimas and maximas

```python
import sympy as sp

# Define the variable and function
x = sp.symbols('x')
f = x**5 - 10*x**3 + 15*x

# First derivative of f(x)
derivative_f = sp.diff(f, x)

# Find critical points by solving f'(x) = 0
critical_points_f = sp.solve(derivative_f, x)

# Convert critical points to decimal form for evaluation
critical_points_f_decimal = [sp.N(point) for point in critical_points_f]

# Compute the second derivative of f(x)
second_derivative_f = sp.diff(derivative_f, x)

# Arrays to hold minimas and maximas
arr_minimas = []
arr_maximas = []

# Classify critical points based on the second derivative
for point in critical_points_f_decimal:
    second_derivative_value = float(second_derivative_f.subs(x, point))

    # If the second derivative is positive, it's a minimum
    if second_derivative_value > 1e-10:
        arr_minimas.append((point, float(f.subs(x, point))))

    # If the second derivative is negative, it's a maximum
    elif second_derivative_value < -1e-10:
        arr_maximas.append((point, float(f.subs(x, point))))

# Print the arrays of minimas and maximas
print(f'the array of minimas {arr_minimas} and the array of maximas is {arr_maximas}')
```

```python
# Find the global maxima and minima by comparing y-values
Global_maxima = max(arr_maximas, key=lambda t: t[1]) if arr_maximas else None
Global_minima = min(arr_minimas, key=lambda t: t[1]) if arr_minimas else None

# Print the global maxima and minima
print(f'The global maxima is {Global_maxima}')
print(f'The global minima is {Global_minima}')
```

# Task 5: Finding the Area under the curve

```python
def area_under_curve(func, var, lower_limit, upper_limit):
    # Define the variable
    var = sp.symbols(var)
    # Compute the definite integral of the function between the limits
    area = sp.integrate(func, (var, lower_limit, upper_limit))
    return area

# Define the variable and function
x = sp.symbols('x')
f = x**5 - 10*x**3 + 15*x

# Define limits for area calculation
lower_limit = -2.3344142183389773
upper_limit = 2.3344142183389773

# Calculate the area under the curve
area = area_under_curve(f, 'x', lower_limit, upper_limit)

# Output the result
print(f"The area under the curve from x = {lower_limit} to x = {upper_limit} is {area}")
```
